```python
import pandas as pd
 
df = pd.read_excel(io = "bne_2015.xlsx", sheet_name="CUADRO3", header =4)
colu_limpias = df.drop(['Unnamed: 0','Unnamed: 7','Unnamed: 8'], axis=1)
ulimas_5 = colu_limpias.drop([0,9,10,11,12,13,14],axis=0)
ano_2015 = ulimas_5.assign(Año=2015)
cambio_nombre = ano_2015.rename(columns={'Consumo Bruto (1)':'Oferta Primaria','Var. Stock+ Perd y Error Est.':'Variación de stock'})
limpieza_2015 = cambio_nombre
limpieza_2015
```

```python
import pandas as pd
 
df = pd.read_excel(io = "bne_2016.xlsx", sheet_name="CUADRO3", header =6)
colu_limpias = df.drop(['Unnamed: 0'], axis=1)
ulimas_5 = colu_limpias.drop([8,9,10,11,12],axis=0)
ano_2016 = ulimas_5.assign(Año=2016)
eliminar_carac = ano_2016.replace('\*','',regex=True)
limpieza_2016 =eliminar_carac
limpieza_2016
```

```python
import pandas as pd
 
df = pd.read_excel(io = "bne_2017.xlsx", sheet_name="CUADRO3", header =6)
colu_limpias = df.drop(['Unnamed: 0'], axis=1)
ulimas_5 = colu_limpias.drop([9,10,11,12,13],axis=0)
ano_2017 = ulimas_5.assign(Año=2017)
eliminar_carac = ano_2017.replace('\*','',regex=True)
limpieza_2017 = eliminar_carac
limpieza_2017
```

```python
df = pd.read_excel(io = "bne_2018.xlsx", sheet_name="CUADRO3", header =6)
colu_limpias = df.drop(['Unnamed: 0'], axis=1)
ulimas_5 = colu_limpias.drop([9,10,11,12,13],axis=0)
ano_2018 = ulimas_5.assign(Año=2018)
eliminar_carac = ano_2018.replace('[#,@,&,*,(,)]','',regex=True)
limpieza_2018 = eliminar_carac
limpieza_2018
```

```python
df = pd.read_excel(io = "bne_2019.xlsx", sheet_name="CUADRO3", header =6)
colu_limpias = df.drop(['Unnamed: 0'], axis=1)
ulimas_5 = colu_limpias.drop([9,10,11,12,13],axis=0)
ano_2019 = ulimas_5.assign(Año=2019)
eliminar_carac = ano_2019.replace('[#,@,&,*,(,)]','',regex=True)
limpieza_2019 =eliminar_carac
limpieza_2019
```

```python
df = pd.read_excel(io = "bne_2020.xlsx", sheet_name="CUADRO3", header =6)
colu_limpias = df.drop(['Unnamed: 0'], axis=1)
ulimas_5 = colu_limpias.drop([9,10,11,12,13],axis=0)
ano_2020 = ulimas_5.assign(Año=2020)
eliminar_carac = ano_2020.replace('[#,@,&,*,(,)]','',regex=True)
limpieza_2020 = eliminar_carac
limpieza_2020
```

```python
tabla_limpieza = pd.concat([limpieza_2016,limpieza_2017,limpieza_2018,limpieza_2019,limpieza_2020,limpieza_2015])
tabla_limpieza
```

```python
#Sumar oferta primaria por tipo de energia
suma_tipo_energia = tabla_limpieza.groupby(['Energético'])['Oferta Primaria'].agg('sum')
suma_tipo_energia
```

```python
#Gráfico 3
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib import colors

# df1 = pd.read_csv("Administracion y Finanzas.csv", sep = ';')

remu = list(tabla_limpieza.groupby(['Energético'])['Oferta Primaria'].agg('sum'))

total = sum(remu)
normdata = colors.Normalize(min(remu), max(remu))
colormap = cm.get_cmap("Blues")
colores =colormap(normdata(remu))

anio = tabla_limpieza['Energético'].unique()
plt.pie(remu, labels=anio, autopct=lambda p: '{:.0f}'.format(p * total / 100), colors=colores)
plt.title(f'Tipo de combustible años 2015-2020', size = 50)
plt.axis("equal")
plt.show()
print(remu)
```

```python
#Librerías necesarias para cargar dataframes a .db
import pandas as pd
import sqlite3

#Creamos archivo .db en el directorio del programa
connection = sqlite3.connect('bd_energia_empresa.db')
c = connection.cursor()

#Creamos la tabla donde se cargará el dataframe
c.execute('CREATE TABLE IF NOT EXISTS consumos (product_name text, price number)')
connection.commit()

#Ejemplo de dataframe creado en pandas en base a una colección del tipo diccionario
# data = {'product_name': ['Computer','Tablet','Monitor','Printer'],'price': [900,300,450,150]}
df = pd.DataFrame(tabla_limpieza, columns= ['Energético','Producción bruta','Importación','Exportación','Variación de stock','Oferta Primaria','Año'])
df.to_sql('consumos', connection, if_exists='replace', index = False)

#Consultamos la tabla creada en la base de datos .db para ver que proceso fue exitoso
df = pd.read_sql('SELECT * FROM consumos', connection)
print(df)

```

```python
import pandas as pd
import glob
# Primero especificamos un patrón del archivo y lo pasamos como parámetro en la función glob
xlsx_files = glob.glob('*.xlsx')
# Mostrar el archivo xlsx_files, el cual es una lista de nombres
print(xlsx_files)
```

```python
list_data = []
  
# Escribimos un loop que irá a través de cada uno de los nombres de archivo a través de globbing y el resultado final será la lista dataframes
for filename in xlsx_files:
    data = pd.read_excel(filename)
    list_data.append(data)
#Para chequear que todo está bien, mostramos la list_data por consola
list_data
pd.concat(list_data,ignore_index=True)

# colu_limpias = conca.drop(['Unnamed: 0','Unnamed: 1','Unnamed: 2'], axis=1)
# colu_limpias
```
